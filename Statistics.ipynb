```python
#Mode
import statistics
data = [1, 2, 2, 3, 4, 4, 4, 5, 6]
mode_value = statistics.mode(data)
print(f"The mode of the dataset is: {mode_value}")
```

```python
#Outlier
data_outliers = [10, 12, 14, 15, 16, 18, 19, 20, 100]  # 100 is an outlier
Q1 = np.percentile(data_outliers, 25)
Q3 = np.percentile(data_outliers, 75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data_outliers if x < lower_bound or x > upper_bound]
print(f"Outliers: {outliers}")
```

```python
#Variance
import numpy as np
import statistics
data=[10,12,13,14,15,16,17,18]
variance_numpy=np.var(data,ddof=0)
variance_numpy_sample=np.var(data,ddof=1)
variance_statistics=statistics.variance(data)
variance_statistics_population=statistics.pvariance(data)
print(f"Variance (NumPy - Population): {variance_numpy}")
print(f"Variance (NumPy - Sample): {variance_numpy_sample}")
print(f"Variance (Statistics - Sample): {variance_statistics}")
print(f"Variance (Statistics - Population): {variance_statistics_population}")
```

```python
#standard deviation
import numpy as np
import statistics
data = [10, 12, 14, 15, 16, 18, 19, 20]
std_dev_numpy = np.std(data, ddof=0)  
std_dev_numpy_sample = np.std(data, ddof=1)  
std_dev_statistics = statistics.stdev(data)  
std_dev_statistics_population = statistics.pstdev(data)  
print(f"Standard Deviation (NumPy - Population): {std_dev_numpy}")
print(f"Standard Deviation (NumPy - Sample): {std_dev_numpy_sample}")
print(f"Standard Deviation (Statistics - Sample): {std_dev_statistics}")
print(f"Standard Deviation (Statistics - Population): {std_dev_statistics_population}")
```

```python
#one tailed z-test hypothesis
import numpy as np
import scipy.stats as stats
sample_mean=100
population_mean=100
population_std=15
sample_size=50
alpha=0.5
z_score=(sample_mean-population_mean)/(population_std/np.sqrt(50))
print('Z-score :',z_score)
z_critical=stats.norm.ppf(1-alpha)
print('critical Z-score :',z_critical)
if z_score > z_critical:
    print("Reject NULL hypothesis")
else:
    print("Fail to reject Null hypothesis")
p_value=1-stats.norm.cdf(z_score)
print('p-value :',p_value)
if p_value <  alpha:
    print("Reject Null Hypothesis")
else:
  print("Fail to Reject Null Hypothesis")
```

```python
#Two sampled z=test
import numpy as np
import scipy.stats as stats
#Group a (offline class)
n1=20
x1=30
y1=40
#Group b (online class)
n2=15
x2=25
y2=35

D=0

alpha=0.05

z_score==((x1-x2)-D)/np.sqrt((y1**2/n1)+(y2**2/n2))
print('Z-score :',np.abs(z_score))
z_critical=stats.norm.ppf(1-alpha/2)
print('Critical z_score: ',z_critical)
if np.abs(z_score)>z_critical:
    print("""Reject the null hypothesis.
There is a significant difference between the online and offline classes.""")
else:
    print("""Fail to reject the null hypothesis.
There is not enough evidence to suggest a significant difference between the online and offline classes.""")

p_value=2*(1-stats.norm.cdf(np.abs(z_score)))
print('P-value :',p_value)
if p_value < alpha:
    print("""Reject the null hypothesis.
There is a significant difference between the online and offline classes.""")
else:
    print("""Fail to reject the null hypothesis.
There is not enough evidence to suggest significant difference between the online and offline classes.""")
```

```python
#visualize the chi-square distribution and Critical Region, with python code.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
df = 2
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, df)
calculated_chi_square = 3.747
x = np.linspace(0, 10, 1000)
y = stats.chi2.pdf(x, df)
plt.plot(x, y, label='Chi-Square Distribution (df=2)')
plt.fill_between(x, y, where=(x > critical_value), color='red', alpha=0.5, label='Critical Region')
plt.axvline(calculated_chi_square, color='blue', linestyle='dashed', label='Calculated Chi-Square')
plt.axvline(critical_value, color='green', linestyle='dashed', label='Critical Value')
plt.title('Chi-Square Distribution and Critical Region')
plt.xlabel('Chi-Square Value')
plt.ylabel('Probability Density Function')
plt.legend()
plt.show()


```

```python
#implementation of Chi-Square feature selection
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
iris=load_iris()
X=iris.data
y=iris.target
column_names=[f'feature_{i}' for i in range (X.shape[1])]
df=pd.DataFrame(X, columns=column_names)
df['target']=y
print("Original Dataset :")
print(df.head())
k=2
chi2_selector=SelectKBest(chi2,k=k)
X_new=chi2_selector.fit_transform(X,y)
selected_features = df.columns[:-1][chi2_selector.get_support()]
print("\nSelected Features:")
print(selected_features)
```
