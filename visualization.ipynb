```python
#Histogram
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(r'C:\Users\preshtha\Documents\dataset\tip.csv')
x=data['total_bill']
plt.hist(x)
plt.title("Tips Dataset")
plt.ylabel("Total bill")
plt.xlabel("Frequency")
plt.show()
```

```python
#Histogram in Matplotlib with Density Plot
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data=np.random.randn(1000)
sns.histplot(data,bins=30,kde=True,color="Pink",edgecolor="red")
plt.xlabel("Values")
plt.ylabel("Density")
plt.title("Histogram in matplotlib with Density Plot")
plt.show()
```

```python
#Multiple Histograms with Subplots
import matplotlib.pyplot as plt
import numpy as np
data1=np.random.randn(1000)
data2=np.random.normal(loc=3,scale=1,size=1000)
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(12,4))
axes[0].hist(data1,bins=30,color="green",edgecolor='black')
axes[0].set_title('Histogram 1')
axes[1].hist(data2,bins=30,color="Pink",edgecolor='black')
axes[1].set_title('Histogram 2')
for ax in axes:
    ax.set_xlabel('Values')
    ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
#Stacked Histogram using Matplotlib
import matplotlib.pyplot as plt
import numpy as np
data1=np.random.randn(1000)
dat2=np.random.normal(loc=3,scale=1,size=1000)
plt.hist([data1,data2],bins=30,stacked=True,color=['cyan','purple'],edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Stacked Histogram using Matplotlib')
plt.legend(['Dataset 1','Dataset 2'])
plt.show()
```

```python
#Plot 2D Histogram (Hexbin Plot) using Matplotlib
import matplotlib.pyplot as plt
import numpy as np
x=np.random.randn(1000)
y=2*x+np.random.normal(size=1000)
plt.hexbin(x,y,gridsize=30,cmap='Blues')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Plot 2D Histogram (Hexbin Plot)')
plt.colorbar()
plt.show()
```

```python
#Box Plot
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data=[np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=True,boxprops=dict(facecolor='Pink'),medianprops=dict(color='skyblue'))
plt.xlabel('Data set')
plt.ylabel('Values')
plt.title('Box plot')
plt.show()
```

```python
#Scatter plot
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv(r'C:\Users\preshtha\Documents\dataset\tip.csv')
x=data['day']
y=data['total_bill']
plt.scatter(x,y)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Scatter Plot for Tips Dataset')
plt.show()
```

```python
#Scatter Plot with Multiple Datasets
import matplotlib.pyplot as plt 
import numpy as np
x1 = np.array([160, 165, 170, 175, 180, 185, 190, 195, 200, 205])
y1 = np.array([55, 58, 60, 62, 64, 66, 68, 70, 72, 74])
x2 = np.array([150, 155, 160, 165, 170, 175, 180, 195, 200, 205])
y2 = np.array([50, 52, 54, 56, 58, 64, 66, 68, 70, 72])
plt.scatter(x1,y1,color="blue",label="Group 1")
plt.scatter(x2,y2,color="Pink",label="Group 2")
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.title('Comparison of Height vs Weight between two groups')
plt.legend()
plt.show()
```
