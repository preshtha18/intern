```python
from transformers import BertTokenizer,BertForSequenceClassification,pipeline
from datasets import load_dataset
```

```python
dataset=load_dataset('imdb')
```

```python
test_data=dataset['test']
```

```python
print(dataset)
```

```python
model_name= 'bert-base-uncased'
model=BertForSequenceClassification.from_pretrained(model_name)
tokenizer=BertTokenizer.from_pretrained(model_name)
```

```python
sentiment_analysis=pipeline('sentiment-analysis',model=model,tokenizer=tokenizer)
```

```python
labels={'LABEL_0':"negative",'LABEL_1':"positive"}
```

```python
for i in range(10):
  review=dataset['test'][i]['text']
  print('Review:\n' ,review)

  sentiment=sentiment_analysis(review)
  sentiment_label = labels[sentiment[0]['label']]
  print(f"Sentiment: {sentiment_label}, Confidence: {sentiment[0]['score']:.4f}\n")
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification
```

```python
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
sentiment_analyzer = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
sample_reviews = dataset['test']['text'][:10]
results = sentiment_analyzer(sample_reviews)

for review, result in zip(sample_reviews, results):
    print(f"Review: {review}\nSentiment: {result['label']} with score {result['score']:.4f}\n")
```
