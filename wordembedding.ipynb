```python
#train Word2Vec on sample Text
import pandas as pd
from gensim.models import Word2Vec
from textblob import TextBlob
data={
    'reviews':[
        "The food was amazing and delicious.",
        "I love the ambiance and the service.",
        "The food was cold and not tasty.",
        "Great place for dinner with friends.",
        "The food was okay, nothing special.",
        "I had a wonderful meal, highly recommend!",
        "The service was slow, but the food was good.",
        "I didn't like the food at all.",
        "The dessert was fantastic!",
        "The food was fresh and flavorful."
    ]
}
df=pd.DataFrame(data)
tokenized_reviews=[review.lower().split() for review in df['reviews']]
model=Word2Vec(sentences=tokenized_reviews,vector_size=100,window=5, min_count=1, workers=4)
model.save("word2Vec_restaurant_reviews.model")
```

```python
#Identify the Top 5 Similar Words to "Food"
from gensim.models import Word2Vec
model=Word2Vec.load("word2Vec_restaurant_reviews.model")
similar_words=model.wv.most_similar('food',topn=5)
print("Top 5 similar words to 'food':")
for word, similarity in similar_words:
    print(f"{word}: {similarity:.4f}")
```

```python
#Sentiment Analysis
from textblob import TextBlob
def analyze_sentiment(review):
    analysis=TextBlob(review)
    if analysis.sentiment.polarity>0:
        return 'Positive'
    elif analysis.sentiment.polarity==0:
        return 'Neutral'
    else:
        return 'Negative'
df['sentiment']=df['reviews'].apply(analyze_sentiment)
print("\nCustomer Feedback Sentiment:")
print(df[['reviews','sentiment']])
```

```python
#word vector one hot encoding
import pandas as pd
import numpy as np
data = {
    'reviews': [
        "The food was amazing and delicious.",
        "I love the ambiance and the service.",
        "The food was cold and not tasty.",
        "Great place for dinner with friends.",
        "The food was okay, nothing special.",
        "I had a wonderful meal, highly recommend!",
        "The service was slow, but the food was good.",
        "I didn't like the food at all.",
        "The dessert was fantastic!",
        "The food was fresh and flavorful."
    ]
}
df = pd.DataFrame(data)
#  Create a Vocabulary
unique_words = set()
for review in df['reviews']:
    unique_words.update(review.lower().split())
vocabulary = sorted(unique_words)
print("Vocabulary:")
print(vocabulary)
#One-Hot Encode the Text
def one_hot_encode(review, vocabulary):
    one_hot_vector = np.zeros(len(vocabulary))
    words = review.lower().split()
    for word in words:
        if word in vocabulary:
            index = vocabulary.index(word)
            one_hot_vector[index] = 1
            
    return one_hot_vector
one_hot_encoded_reviews = np.array([one_hot_encode(review, vocabulary) for review in df['reviews']])
print("\nOne-Hot Encoded Vectors:")
for i, review in enumerate(one_hot_encoded_reviews):
    print(f"Review {i+1}: {review}")
```

```python
#CBOW
import pandas as pd
from gensim.models import Word2Vec
data = {
    'reviews': [
        "The food was amazing and delicious.",
        "I love the ambiance and the service.",
        "The food was cold and not tasty.",
        "Great place for dinner with friends.",
        "The food was okay, nothing special.",
        "I had a wonderful meal, highly recommend!",
        "The service was slow, but the food was good.",
        "I didn't like the food at all.",
        "The dessert was fantastic!",
        "The food was fresh and flavorful."
    ]
}
df = pd.DataFrame(data)
tokenized_reviews=[review.lower().split() for review in df['reviews']]
print("Tokenized Reviews:")
print(tokenized_reviews)
#Train the CBOW model
model=Word2Vec(sentences=tokenized_reviews,vector_size=100,window=2,min_count=1,sg=0,workers=4)
model.save("cbow_restaurant_reviews_model")
word_vector=model.wv['food']
print("\n Word vector for 'food':")
print(word_vector)
similar_words = model.wv.most_similar('food', topn=5)
print("\nTop 5 similar words to 'food':")
for word, similarity in similar_words:
    print(f"{word}: {similarity:.4f}")
```
