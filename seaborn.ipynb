```python
#seaborn.FactGrid()
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
g = sns.FacetGrid(tips, col="day", hue="sex", height=4, aspect=1)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()
```

```python
#seaborn.PairGrid()
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
g=sns.PairGrid(iris,hue="species")
g.map_upper(sns.scatterplot)
g.map_diag(sns.histplot,kde=True)
g.map_lower(sns.kdeplot)
g.add_legend()
plt.show()



```

```python
#Scatter plot 
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset("tips")
plt.figure(figsize=(10,6))
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="day",style="sex",s=100)
plt.title("scatter plot of total bill vs tip")
plt.xlabel("Total bill($)")
plt.ylabel("tips($)")
plt.show()
```

```python
# line plot
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
average_sepal_length = iris.groupby("species")["sepal_length"].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_sepal_length, x="species", y="sepal_length", marker="o")
plt.title("Average Sepal Length of Iris Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()
```

```python
#Bar plot
# import module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
State = ["Rajasthan", "Bihar", "Madhya Pradesh",
		"Gujarat", "Maharashtra"]
growth = [342239, 94163, 308245, 196024, 307713]
df = pd.DataFrame({"State": State,
				"Growth": growth})
df.sort_values('Growth')
sns.barplot(x='State',
			y="Growth", data=df, 
			order=df.sort_values('Growth').State)
```
