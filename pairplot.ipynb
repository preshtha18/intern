```python
import matplotlib.pyplot as plt
import seaborn
df=seaborn.load_dataset('tips')
df.head()
```

```python
seaborn.pairplot(df)
plt.show()
```

```python
#Plotting selected Variables
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
selected_vars=['total_bill','tip']
sns.pairplot(df,vars=selected_vars)
plt.show()
```

```python
#Adding a Hue Color to a Seaborn Pairplot
import seaborn as sns
import matplotlib.pyplot as plt
df=seaborn.load_dataset('tips')
seaborn.pairplot(df,hue='size')
plt.show()
```

```python
#Modifying Color Palette
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
sns.pairplot(df,hue='size',palette='husl')
plt.show()
```

```python
#Diagonal Kind of plots
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
sns.pairplot(df,diag_kind='kde')
plt.show()
```

```python
#Adjusting Plot Kind
sns.pairplot(df,kind='reg')
plt.show()
```

```python
#Controlling the Markers
sns.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
#Limiting the Variables
sns.pairplot(df,hue='sex',vars=['total_bill','tip','size'])
plt.show()
```

```python
#Advanced Customization With Seaborn Pairplot and remove tick label
import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('tips')
g=sns.pairplot(df,hue='day')
g.fig.suptitle("Pairplot of Tips Dataset",y=1.02)
for ax in g.axes.flatten():
    ax.set_xticks([])
    ax.set_yticks([])
plt.show()
```
